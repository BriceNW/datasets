# Game of Thrones

Êtes-vous un fan de la série Game of Thrones ? Testons cela. Utilisons l’API : [https://anapioficeandfire.com/](https://anapioficeandfire.com/) pour mieux connaître ce qu’il y a à savoir sur la série

0. Importez les librairies suivantes : 

      * ```requests```
      * ```pandas```
      * ```json```

````python
import requests
import pandas as pd
import json
````

1. Allez sur l’URL indiquez dans la description ci-dessus. En utilisant requests, obtenez les données sur un personnage de la série. Peu importe le personnage

````python
r = requests.get("https://anapioficeandfire.com/api/characters/583")
r.text
````

2. Extrayez ces données sous format JSON

````python
r.json()
````

3. Extrayez maintenant les clés du JSON que vous avez obtenu et créez un DataFrame qui aura pour colonne les clés du JSON que vous avez extrait.

        * Indice, on pourra utiliser:

  ```
  r.json().keys()
  ```

````python
col_names = [key for key in r.json().keys()]
df = pd.DataFrame(columns=col_names)
df
````

4. Créez une boucle pour extraire les données des 150 premiers personnages de la série et ajoutez les à votre DataFrame

````python
for i in range(1, 151):
# url = "https://anapioficeandfire.com/api/characters/{}".format(i)
  url = "https://anapioficeandfire.com/api/characters/" + str(i+1)
  r = requests.get(url)
  data = r.json()
  df = df.append(data, ignore_index=True, sort=False)
  print('character number', i, 'done...')
````
